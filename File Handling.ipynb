```python
import csv
```

```python
f=open('firstcsvfile.csv','w')
w=csv.writer(f)
w.writerow(['NAME','ROLL.NO','MARKS'])
#to dynamically write from keyboard
while True:
    name=input("enter name: ")
    rollno=int(input("enter roll.no: "))
    marks=int(input("enter marks: "))
    w.writerow([name,rollno,marks])
    #if u want to stop the loop
    option=input("do u want to continue yes/no: ")
    if option.lower()=='no':
        break
f.close()
```

```python
#add newline='' if u dont want a empty line in the file 
f=open('firstcsvfile.csv','a',newline='')
w=csv.writer(f)
w.writerow(['NAME','ROLL.NO','MARKS'])
#to dynamically write from keyboard
while True:
    name=input("enter name: ")
    rollno=int(input("enter roll.no: "))
    marks=int(input("enter marks: "))
    w.writerow([name,rollno,marks])
    #if u want to stop the loop
    option=input("do u want to continue yes/no: ")
    if option.lower()=='no':
        break
f.close()
```

```python
f=open("exptcsvfile.csv",'w')
#we have to create a writer object.
wrt=csv.writer(f)
wrt.writerows(['NAME','ROLL','ADDRESS'])
f.close()
```

```python
f=open('firstcsvfile.csv','r')
print(f)
#we have to create a reader object
r=csv.reader(f)
print(r)
#print(list(r))
for i in r:
    print(i)
f.close()
```

```python
f=open('expt3csvfile.csv','w',newline='')
w=csv.writer(f)
fr=open('firstcsvfile.csv','r')
r=csv.reader(fr)
for i in r:
    if i==[]:
        continue
    else:
        w.writerow(i)
f.close()
fr.close()
```

```python
f=open('expt3csvfile.csv','r+',newline='')
w=csv.writer(f)
r1=csv.reader(f)
fr=open('firstcsvfile.csv','r')
r=csv.reader(fr)
for i in r:
    print(list(r1))
    if i==[] and i in r1:
        continue
    else:
        w.writerow(i)
f.close()
fr.close()
```
